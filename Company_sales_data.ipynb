```python
#importing nesccary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#reading the file
df = pd.read_csv("company_sales_data.csv")
df
```

```python
#Question 1: Read Total profit of all months and show it using a line plot. 
profit_by_month = df[['month','total_profit']]
profit_by_month
```

```python
#Line plot
x = df.month
y= df.total_profit
plt.plot(x, y)
plt.show()
```

```python
#Question2: Read data for Bathing soap and facewash of all months and display it using the Subplot.
sales_by_months = df[['month','bathingsoap','facewash']]
sales_by_months
```

```python
#display using subplot
sales = pd.DataFrame(sales_by_months)
x_pos = np.arange(len(df['month']))
fig, ax = plt.subplots(figsize=(12,5))
p1 = ax.bar(x_pos,sales['bathingsoap'], label='Bathing Soap')
p2 = ax.bar(x_pos,sales['facewash'],  label='Facewash')
#showing legends and labels
ax.set_xticks(x_pos, labels=['Jan','Feb','Mar','Apr','May','June','July','August','September','October','November','December'])
ax.legend()
```
